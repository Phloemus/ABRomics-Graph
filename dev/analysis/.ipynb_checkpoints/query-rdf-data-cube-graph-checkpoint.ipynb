# Quering the RDF data cube (QB) graph

```python
# imports

import json
import time
import pandas as pd

from rdflib import Graph
import uuid
from rdflib import Namespace
from rdflib import FOAF, DCTERMS, SKOS, VANN, TIME, RDF, RDFS, XSD, OWL, Literal
```

```python
g = Graph()
g.parse('/home/brieuc/Documents/dev/abromics/abromics-kg/dev/analysis/rdf-data-cube-graph.ttl', format='ttl')
```

```python
## Q1 of hackmd
## Get all the resistance gen for a sample

query = """
    PREFIX qb: <http://purl.org/linked-data/cube#>
    PREFIX schema: <https://schema.org/>
    PREFIX abromics: <https://abromics.fr/>
    PREFIX rdf: <http://www.w3.org/1999/02/22-rdf-syntax-ns#>
    PREFIX rdfs: <http://www.w3.org/2000/01/rdf-schema#>
    PREFIX xsd: <http://www.w3.org/2001/XMLSchema#>
    PREFIX obo: <http://purl.obolibrary.org/obo/>
    PREFIX aro: <http://purl.obolibrary.org/obo/aro.owl#>
    PREFIX go: <http://purl.org/obo/owl/GO#>
    PREFIX uniprot: <http://purl.uniprot.org/uniprot/>
    PREFIX geo: <http://www.w3.org/2003/01/geo/wgs84_pos#>

    SELECT DISTINCT ?resGeneName
    WHERE {
        ?sample schema:identifier "0001" ;
           rdf:type sio:001050 .
           
        ?sampleProperty rdf:type rdf:Property ;
            rdf:type qb:DimensionProperty ;
            rdfs:label "Samples"^^xsd:string .
            
        ?resGeneNameProperty rdf:type rdf:Property ;
            rdf:type qb:DimensionProperty ;
            rdfs:label "Antibiotic resistant gene names"^^xsd:string .
            
        ?dataset rdf:type qb:DataSet .
        ?observations qb:dataSet ?dataset ;
            ?sampleProperty ?sample .
            
        ?observations ?resGeneNameProperty ?resGeneName .
    }
"""

res = g.query(query)

for row in res:
    print(f"{row.resGeneName}")
```

```python
# Q2: List the TOP-K antibiotic resistance genes by coverage

query = """
    PREFIX qb: <http://purl.org/linked-data/cube#>
    PREFIX schema: <https://schema.org/>
    PREFIX abromics: <https://abromics.fr/>
    PREFIX rdf: <http://www.w3.org/1999/02/22-rdf-syntax-ns#>
    PREFIX rdfs: <http://www.w3.org/2000/01/rdf-schema#>
    PREFIX xsd: <http://www.w3.org/2001/XMLSchema#>
    PREFIX obo: <http://purl.obolibrary.org/obo/>
    PREFIX aro: <http://purl.obolibrary.org/obo/aro.owl#>
    PREFIX go: <http://purl.org/obo/owl/GO#>
    PREFIX uniprot: <http://purl.uniprot.org/uniprot/>
    PREFIX geo: <http://www.w3.org/2003/01/geo/wgs84_pos#>

    SELECT DISTINCT ?resGeneName ?coverage
    WHERE {     
        ?resGeneNameProperty rdf:type rdf:Property ;
            rdf:type qb:DimensionProperty ;
            rdfs:label "Antibiotic resistant gene names"^^xsd:string .
            
        ?coverageProperty rdf:type rdf:Property ;
            rdf:type qb:MeasureProperty ;
            rdfs:label "Antibiotic resistant gene coverage"^^xsd:string .
            
        ?dataset rdf:type qb:DataSet .
        ?observations qb:dataSet ?dataset .
            
        ?observations ?resGeneNameProperty ?resGeneName .
        ?observations ?coverageProperty ?coverage .
    }
    ORDER BY DESC(?coverage)
    LIMIT 10
"""

res = g.query(query)

for row in res:
    print(f"{row.resGeneName} - {row.coverage}")
```
