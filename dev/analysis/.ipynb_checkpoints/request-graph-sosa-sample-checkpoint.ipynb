```python
# imports

from rdflib import Graph
import uuid
from rdflib import Namespace
from rdflib import FOAF, DCTERMS, SKOS, VANN, TIME, RDF, RDFS, XSD, OWL, Literal
```

```python
g = Graph()
g.parse('/home/brieuc/Documents/dev/abromics/abromics-kg/dev/analysis/antibiotic-resistance-sosa-sample.ttl', format='ttl')
```

```python
query = """
    PREFIX ns1: <http://www.abromics.org/ontology/>
    PREFIX rdfs: <http://www.w3.org/2000/01/rdf-schema#>
    PREFIX skos: <http://www.w3.org/2004/02/skos/core#>
    PREFIX sosa: <http://www.w3.org/ns/sosa/>
    PREFIX ssn: <http://www.w3.org/ns/ssn/>
    PREFIX xsd: <http://www.w3.org/2001/XMLSchema#>
    PREFIX prov: <http://www.w3.org/ns/prov#>
    PREFIX aro: <http://purl.obolibrary.org/obo/aro.owl#> # Here we deleted the aro.owl part of the url
    
    SELECT DISTINCT ?subClasses ?wSubClasses 
    WHERE {
        ?sample a sosa:Sample . 
        ?sample sosa:isSampleOf ns1:Metagenomic .
        ?workflow sosa:hasInput ?sample .
        ?workflow rdf:type ?wSubClasses .
        ?workflow sosa:hasOutput ?report .
        ?report ns1:contains ?content . 
        ?content rdf:type ?subClasses .
    }
"""

res = g.query(query)

##! Find a way to load the aro ontology. As no one hosts it, we should add it to the Virtuoso graph used used for IBD-KG
##! There is a way to add a graph file (in remote server at workfdir/abromics-rdf) in virtuoso : put the graph file in 
##! workdir/abromics-rdf and there is something else to do to notify virtuoso to add the graph file too

for row in res:
    print(f"{row.subClasses} - {row.wSubClasses}")
```
