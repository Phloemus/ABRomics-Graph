# GeoSPARQL graph

The goal of this notebook is to create understand how to use geosparql to filter points in a defined area having geo spatial coordinates

:warn: RDF lib package in python doesn't support GeoSPARQL. The test should be performed on an instanced deployed using a virtuoso graph server.

```python
# imports

import json
import time
import pandas as pd

from rdflib import Graph
import uuid
from rdflib import Namespace
from rdflib import FOAF, DCTERMS, SKOS, VANN, TIME, RDF, RDFS, XSD, OWL, Literal
```

```python
g = Graph()
g.parse('/home/brieuc/Documents/dev/abromics/abromics-kg/dev/analysis/geosparql-graph.ttl', format='ttl')
```

```python
# Query to check if a point is in a certain area with the point geo spacial coordinates

query = """
    SELECT (count(*) as ?c) WHERE {
        ?point rdf:type sf:Point .
        ?poly rdf:type sf:Polygon .
        # FILTER(geof:sfWithin(?poly, ?point))
        ?point geo:sfWithin ?poly .
    }
"""

res = g.query(query)

for row in res:
    print(f"{row.c}")
```
