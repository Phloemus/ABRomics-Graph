```python
## Call the BigsDB API
```

```python
## List of species present in BigDB

response = requests.get("https://rest.pubmlst.org/db")
response.json()
for i in range(0, len(response.json())):
    print(f"{response.json()[i]['name']}")
```

```python
## List of some databases disponible in bigsdb (for certain species there may be multiple databases)

response = requests.get("https://rest.pubmlst.org/db")
response.json()
for i in range(0, len(response.json())):
    print(f"{response.json()[i]['databases'][0]['href']}")
```

```python
## Getting all the loci for the first database containing data about acinetobacter baumanii 

response = requests.get("https://rest.pubmlst.org/db/pubmlst_abaumannii_isolates/loci")
for lociUrl in response.json()['loci']:
    print(f"{lociUrl}")
```

```python
## Getting all the informations about the ACNI00005 locus of the first database containing data about acinetobacter baumanii
#! Problem !# It's impossible to access the alleles of the locus, maybe this specific locus doesn't have any alleles ? 

response = requests.get("https://rest.pubmlst.org/db/pubmlst_abaumannii_isolates/loci/ACIN00005/alleles")
print(response.json())
```

```python
## Trying the same thing using pubmd db instead of the isolates db. Some species don't have any seqdef db

response = requests.get("https://rest.pubmlst.org/db/pubmlst_achromobacter_seqdef/loci")
print(response.json())
```

```python
## Possible to grab the alleles for the rpoB loci. So the data of the allele require to use a seqdef database (ex: here it's "pubmlst_achromobacter_seqdef")
#? By the way, rpoB is a Pasteur MLST locus

response = requests.get("https://rest.pubmlst.org/db/pubmlst_achromobacter_seqdef/loci/rpoB/alleles")
print(response.json())
```
