# ABRomics report extraction from ABRomics API

The ABRomics API allows to fetch ABRomics reports directly with an http request. To have access to your personal reports, enter your password and user name in the .env file (at the root of the project). The following code extract your ABRomics credentials and perform a try to fetch the reports you own.

## Loading the user credentials

Use the content of the **.env** file to retreive user credentials. This will allow to perform requests to the ABRomics API using the permission level of the user registered

```python
## env variables (get the user ABRomics credentials)
load_dotenv() #! when the .env values change there is no reload it takes the old value... 

## load the environment variables
api_url = os.getenv('API_URL') #! not the right value..
api_username = os.getenv('API_USERNAME')
api_password = os.getenv('API_PASSWORD')
api_temp_basic_token = os.getenv('API_TEMP_BASIC_TOKEN') # will be changed at one point find a way to get a new token via a login with creds
```

 ## Downloading reports

Download the first page of reports of the authenticated user. Get the ids of the analysis that yield a report (tagged with the status "ready_to_report") and download them locally into the reports folder 

TODO: Read the report files and complete the SOSA graph using them (the organization of the data in the report were changed a lot)

```python
## get the analysis ids of every analysis that fully worked (has a ready_to_report) status
response = requests.get(
    "https://analysis.abromics.fr/api/analysis/",
    params={'status': 'ready_to_report'},
    headers = {
        'Authorization': f"Basic {api_temp_basic_token}",
    }
)
exploitable_analysis = response.json()
exploitable_analysis
exploitable_analysis_ids = [analysis["id"] for analysis in exploitable_analysis["results"]]
exploitable_analysis_ids
```

```python
## create the reports folder and populate it with report from the ABRomics API

if not os.path.exists("reports"):
    os.makedirs("reports")
    
with alive_bar(len(exploitable_analysis_ids)) as bar:
    for report_id in exploitable_analysis_ids:
        print(report_id)
        response = requests.get(
            f"https://analysis.abromics.fr/api/analysis/{report_id}/report/",
            headers = {
                'Authorization': f"Basic {api_temp_basic_token}",
            }
        )
        report = response.json()
        
        # save the reports in a local file
        with open(f'reports/abr_report_{report_id}.json', 'w') as f:
            json.dump(report, f)
        bar()
```
