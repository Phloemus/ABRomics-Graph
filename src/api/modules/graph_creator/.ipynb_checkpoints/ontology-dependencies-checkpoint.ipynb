```python
## This notebook allows to investigate on the ontologies dependencies
## and the classes that are reused a lot among the ontologies used 
## in the abromics project
```

```python
## imports
from rdflib import Graph
from SPARQLWrapper import SPARQLWrapper, JSON
```

```python
## load two ontologies in a small graph
g = Graph()
g.parse("../../ontologies/aro.owl", format="xml")
g.parse("../../ontologies/ohmi.owl", format="xml")
```

```python
## Total number of nodes in the graph
nb_nodes = len(g)
print(f"Total number of nodes in the graph: {nb_nodes}")
```

```python
## Number of distinct nodes in the graph
count_distinct_nodes = """
SELECT (COUNT(DISTINCT ?s) AS ?count)
WHERE {
  ?s ?p ?o.
}
"""
```

```python
result = g.query(count_distinct_nodes)
print(f"Number of distinct nodes in the graph: {int(result.bindings[0]['count'])}")
```

```python
####################### Finding dependencies between ontologies #######################
# 
# Finding the classes that are common among different ontologies and linking getting 
# their classes of origin.
# 
# A parent classes may have many children in other ontologies which can be investigated
# using the subClassOf keyword
#
```

```python
## Dependency query
dependency_query = """
    PREFIX rdf: <http://www.w3.org/1999/02/22-rdf-syntax-ns#>
    PREFIX rdfs: <http://www.w3.org/2000/01/rdf-schema#>
    PREFIX ibdb: <http://www.ibdb.org/ontology/>


    # Select all the classes having a relation subClassOf with the origin class
    SELECT ?class ?superClass
    WHERE {
      ?class rdfs:subClassOf ?superClass .
      FILTER(STRSTARTS(STR(?superClasse), STR(ibdb:)) = false) . ## will make the query not work but should be useful at one point
    }
"""
```

```python
results = g.query(dependency_query)
print(len(results))
for result in results:
    origin_class, dependence_class = result
    print(f"Origin class : {origin_class}, Dependence class : {dependence_class}")
```
