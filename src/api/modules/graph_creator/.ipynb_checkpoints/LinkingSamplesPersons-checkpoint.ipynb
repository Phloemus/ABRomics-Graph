# Linking Samples and Persons

```python
samples = """
"""

persons = """
"""

attribution = """
"""
```

```python
initial_links = [
    ("Sample#6","Person#1"), 
    ("Sample#6","Person#2"), 
    ("Sample#2","Person#3"), 
    ("Sample#3","Person#3"), 
    ("Sample#4","Person#4"), 
]
```

```python
prefix = "http://abromics.fr/ontology/"
for i in range(1,5):
    p = f"<{prefix}Person#{i}> rdf:type foaf:Person .\n"
    persons += p
    s = f"<{prefix}Sample#{i}> rdf:type abr:Sample .\n"
    samples += s
```

```python
for link in initial_links:
    l = f"<{prefix}{link[0]}> prov:wasAttributedTo <{prefix}{link[1]}> .\n"
    attribution += l
    l = f"<{prefix}{link[1]}> prov:used <{prefix}{link[0]}> .\n"
    attribution += l
```

```python
print(attribution)
print(samples)
print(persons)
```

```python
from rdflib import ConjunctiveGraph
kg = ConjunctiveGraph()

prefixes = """
@prefix prov: <http://prov/> .
@prefix abr: <http://abromics/> .
@prefix rdf: <http://rdf/> .
@prefix foaf: <http://foaf/> .
"""

kg.parse(data=prefixes+samples, format="ttl")
kg.parse(data=prefixes+persons, format="ttl")
kg.parse(data=prefixes+attribution, format="ttl")
print(kg.serialize(format="ttl"))
```
