## ARO annotation of ABRomics results

The goal is to create a small graph that links important informations present in the ABRomics report to terms of the ARO ontology allowing to enrich the data even more

```python
# imports

import json
import time
import pandas as pd

from rdflib import Graph
import uuid
from rdflib import Namespace
from rdflib import FOAF, DCTERMS, SKOS, VANN, TIME, RDF, RDFS, XSD, OWL, Literal
```

```python
# minimal data from an ABRomics report
# this is the kind of data expected to be annotated with ARO

resistanceGene = "aadA13"
targetAntibiotic = "streptomycin"
```

```python
g = Graph()

# Placeholder namespaces
ABROMICS = Namespace("http://www.abromics.org/ontology/")
ARO = Namespace('https://bioportal.bioontology.org/ontologies/ARO/')

## Resistance Gene
resistanceGeneId = uuid.uuid1()
g.add((ABROMICS[str(resistanceGeneId)], RDF.type, ARO.aadA13))

## Target Antibiotic
targetAntibioticId = uuid.uuid1()
g.add((ABROMICS[str(targetAntibioticId)], RDF.type, ARO.streptomycin))

g.serialize(destination="aro-annotation.ttl", format='ttl')
    
g
```
