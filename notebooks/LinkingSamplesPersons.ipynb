# Linking Samples and Persons

```python
import uuid

index_samples = {}

def genSample(j):
    uri = ""
    if j in index_samples.keys():
        uri = index_samples[j]
    else:
        uri = f"{prefix}Sample#{uuid.uuid1()}"    
    
    s = f"""<{uri}> rdf:type abr:Sample ; schema:identifier "{j}".\n"""
    index_samples[j] = uri
    return s,uri

def genPerson():
    i = uuid.uuid1() 
    uri = f"{prefix}Person#{i}"
    s = f"<{prefix}Person#{i}> rdf:type foaf:Person .\n"
    return s,uri

#print(genSample())
#print(genSample())
#print(genPerson())
```

```python
_,uri1 = genPerson()
sample ,uri2 = genSample("R0065")
print(f"{uri1} --used--> {uri2}")
print(sample)
print(index_samples)
```

```python
#assert that we reuse the URI when identifiers are the same
sample1 ,uri1 = genSample("R0065")
sample2 ,uri2 = genSample("R0065")
sample3 ,uri3 = genSample("R0123")

print(sample1, uri1)
print(sample2, uri2)
print(sample3, uri3)

assert uri1 == uri2
assert uri1 != uri3
assert uri2 != uri3
```
